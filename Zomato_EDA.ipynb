```python
df_c = pd.read_excel("EDA_data/Country-Code.xlsx")
df_c.head()
```

we have two table, `"country code"` is common column<br>
now prefom left join

```python
final_df = pd.merge(df,df_c, on="Country Code", how="left")
final_df.head()
```

```python
# count how many country are there
final_df.Country.value_counts()
```

```python
# number of country
n_country = final_df.Country.value_counts().index
# n_country = final_df.Country.nunique()
n_country
```

```python
v_country = final_df.Country.value_counts().values
v_country
```

```python
plt.pie(v_country, labels=n_country,autopct='%0.1f%%')
plt.show()
```

```python
# only top 3 countries
plt.pie(v_country[:3], labels=n_country[:3],autopct='%0.1f%%')
plt.show()
```

`Observation`: Zomato `maximum` transaction are from `INDIA` (90.6) followed by `UK` and `USA`

```python
final_df.columns
```

```python
rating_df = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0 : "Rating count"})
rating_df
# reset_index() to reset index,
# renam() to rename column
```

```python
sns.countplot(x="Rating color", data=rating_df, hue= 'Rating color',
              palette=['cyan','red','orange','yellow','green','darkGreen'])
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating', y='Rating count', data=rating_df, hue='Rating color',
            palette=["cyan","red","orange","yellow","green","darkGreen"])
```

```python
rating_df["Rating count"].sum()
```

`Observation`: maximum people rated between `2.6` to `4.1`

```python
# Find the country where rating is 0
final_df[final_df['Aggregate rating']== 0.0].groupby(['Country']).size().reset_index()
```

`Observation`: maxiumum number of 0 rating from `INDIA`

```python
# which country have online delivery
final_df[["Country","Has Online delivery"]].groupby(["Has Online delivery","Country"]).size().reset_index()
```
